## Intro to github

In this notebook, you'll follow instructions for tracking your code in a git repository

**Git** is the version control tool; you have git installed locally and it stores metadata about the history of your changes.

**Github** is the platform where we and others upload our code and the metadata about the history of our changes, so you can view others' changes. It is a remote server where we can save our changes and view others' code changes. It also has useful project management tools, like issues.

You can confirm that you have git installed by typing "git" into the cmd line. If you've cloned this repository following the instructions, you have git installed.

Confirm that you have a github account by going to this site and registering: https://github.com/

### Workflow

Source: https://gist.github.com/CiceroLino/901b9cf6448bb98bba28060762b1c04e

1) Create a branch in this repository and name it after yourself:

`git checkout -b "your name"`

Now, the changes that you make will not be made to the "main" branch, but will be applied only to your branch. Once you have something that you're happy with, you can merge it into main. This allows you to make changes without worrying that you'll break the "main" version. It also gives you time to test your changes, seek code review, and verify that you're happy with your work.

2. See what the difference is between your environment and the "tracked version" (changes that have been committed) with the command:

`git status`

What does git status tell you?

3. Add your single cell data structures work (to be committed)

`git add 00_scRNA_data_structures/00_data_structures.ipynb`

4. Now, you have a file with tracked changes that you can commit to the repository. Remember, a commit is a snapshot of the code, so you when you commit code you are saving the state of the code. This is useful to do after you make a discreet set of changes, for instance, if I write code to create a new plot or add a feature to my tool, I track the changed files (with git add) and commit then. Always make your commit message meaningful! Something like "track new file" tells you nothing about the history of the changes, so if you want to go back and understand what you were doing, it's not useful.

`git commit -m "commit message"`

5. See what the difference is between your environment and the "tracked version" with the command:

`git status`

What does git status tell you?

6. Now, push your changes to GitHub so that they're preserved on the remote server and your colleagues can see them:

`git push` --set-upstream origin "my branch name"

Now! Set up a new repository to track your current project. You can do this in the github interface (Repositories --> New repository)
